```python
#蛮力提取4000行数据出来，计算每个经纬度附近的16个房价
data=[[]]

#len(df['coordinate_x'])
for i in range(400):
    print(i)
    n=1
    
    for j in range(400):
        if(n<=16):
            if df['coordinate_x'][i]-df['coordinate_x'][j] < 0.0001 and df['coordinate_x'][i]-df['coordinate_x'][j]>-0.0001: 
                
                data[i].append(df['price'][j])
                n=n+1
               
        else:
            break
    data.append([])
```

```python
data = pd.read_csv('round_price_data1.csv')

```

```python
#算数平均每行数据，因为有的周边房价不全，只在数据集中某坐标范围内获取了不足16个数据
row_mean=data.mean(axis=1)#计算每行的平均值
print(type(row_mean))
for i,v in row_mean.items():
    #填补每行空的数据
    data.loc[i]=data.loc[i].fillna(v)
print('data:',data)
data.to_csv('round_price_data1.csv',index=False)
```
