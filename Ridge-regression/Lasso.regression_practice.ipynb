```python
from sklearn import linear_model
import pandas as pd
```

```python
df=pd.read_csv('../xg_boost/ann_train_data.csv')
df=df.drop(['Unnamed: 0'], axis = 1)
df
```

```python
train_dataset = df.sample(frac=0.8,random_state=0)
test_dataset = df.drop(train_dataset.index)
train_dataset
test_dataset 
```

```python
train_labels = train_dataset.pop('price')/10000000
test_labels = test_dataset.pop('price')/10000000
```

```python
ridge = linear_model.Lasso(alpha=1.0, fit_intercept=False,selection='random')
ridge.fit(train_dataset, train_labels)
```

```python
example_result=ridge.predict(test_dataset)
print(example_result)
```

```python
print(example_result[:]*10000000)
print((example_result[:]*10000000-test_labels*10000000)<200000)
j=0
for i in ((example_result[:]*10000000-test_labels*10000000)<200000):
    if(i==True):
        j=j+1
```

```python
print(j)
print(j/len(example_result))
```
