```python
import pandas as pd
import pathlib
import seaborn as sns
```

```python
#df=pd.read_csv('o_train_data.csv')
df=pd.read_csv('ann_train_data.csv')
df=df.drop(['Unnamed: 0'], axis = 1)
df.dtypes
```

```python
train_dataset = df.sample(frac=0.8,random_state=0)
test_dataset = df.drop(train_dataset.index)
train_dataset
test_dataset 
```

```python
train_labels = train_dataset.pop('price')/10000000
test_labels = test_dataset.pop('price')/10000000
print(train_dataset.dtypes)
print(test_labels.dtypes)
len(train_dataset.keys())
```

```python
from tensorflow import keras
from tensorflow.keras import layers
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
  ])
```

```python
model.summary()
```

```python
#model.compile(loss="mean_squared_error", optimizer="sgd", metrics=["accuracy"])
import tensorflow as tf
optimizer = tf.keras.optimizers.RMSprop(0.001)
model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
```

```python
history=model.fit(train_dataset, train_labels,batch_size=64, epochs=10,validation_split = 0.2,verbose=1)
```

```python
example_batch = test_dataset[:10]
example_result = model.predict(example_batch)
print(example_result)
train_labels[:10]
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
import matplotlib.pyplot as plt
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch']*100000, hist['mae']*100000,
           label='Train Error')
  plt.plot(hist['epoch']*100000, hist['val_mae']*100000,
           label = 'Val Error')
  plt.ylim([0,5])
  plt.legend()

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch']*100000, hist['mse']*100000,
           label='Train Error')
  plt.plot(hist['epoch']*100000, hist['val_mse']*100000,
           label = 'Val Error')
  plt.ylim([0,20])
  plt.legend()
  plt.show()
```

```python
plot_history(history)
```
