```python
import numpy as np
from sklearn.svm import SVR
import pandas as pd
```

```python
df=pd.read_csv('../xg_boost/ann_train_data.csv')
df=df.drop(['Unnamed: 0'], axis = 1)
df
```

```python
train_dataset = df.sample(frac=0.8,random_state=0)
test_dataset = df.drop(train_dataset.index)
train_dataset
test_dataset 
```

```python
train_labels = train_dataset.pop('price')/10000000
test_labels = test_dataset.pop('price')/10000000
```

```python
#CCC
#gammagamma
#https://blog.csdn.net/wusecaiyun/article/details/49681431
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
```

```python
model=svr_rbf.fit(train_dataset, train_labels)
```

```python
example_result =model.predict(test_dataset)
print(example_result[:]*10000000)
print((example_result[:]*10000000-test_labels*10000000)<200000)
j=0
for i in ((example_result[:]*10000000-test_labels*10000000)<200000):
    if(i==True):
        j=j+1
```

```python
print(j)
print(j/len(example_result))
```
